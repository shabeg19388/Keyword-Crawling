```python
#Shabeg Singh Gill
#IIITD

import pandas as pd 
import requests 
import numpy as np
from bs4 import BeautifulSoup as b

lenovo=[] #Lenovo laptop list , finding keyword Lenovo to filter only Lenovo laptops on amazon 
asus=[] #ASUS laptop list, finding keyword ASUS to filter only lenovo laptops 
url=[] #list of urls containing the keywords 
#iterating through pages 
for x in range(1,8):
    #requesting for url 
    mainurl="https://www.amazon.in/s?k=laptops&page="
    #requesting for url
    req= requests.get(mainurl+str(x))
    #parsing using html/lxml
    cururl=mainurl+str(x)
    #print(cururl)
    soup =b(req.content,'html.parser')
    #searching for span tags which contains laptop names
    lap= soup.find_all('span', {'class':'a-size-medium a-color-base a-text-normal'})
    #print(lap)
    #iterating over span tags i.e. laptop names 
    check=0 #for checking url signficance status 
    for i in lap: 
        
        #checking if laptop name is null or not 
        if i.text is not None: 
            
            #checking for lenovo(keyword) laptops
            if 'Lenovo' in i.text: #if laptop is of brand lenovo, then adding to list 
                name=i.text #getting only text out of span 
                #print(name) #printing 
                #print(" ") #next line 
                lenovo.append(name) #appending to list 
                check=1 #changing to 1 if lenovo present, so url should be added to csv 
    
    if (check==1) :
        #print(mainurl+str(x))
        url.append(cururl)
    
#csv file-1 for relevant urls only
b= {'URL':url}
df1=  pd.DataFrame.from_dict(b, orient='index') 
df1=df1.transpose() 
print(df1)
df1.to_csv('URL.csv') #csv file-1 

print(" ")
#csv file-2 for relevant laptops(keywords) only 
a= {'Lenovo Laptops': lenovo}  
df2= pd.DataFrame.from_dict(a, orient='index') 
df2=df2.transpose() 
print(df2)
df2.to_csv('Lenovo.csv') #csv file-1 
          
#Also, we will have two CSV files:
#1. For URLs having data with the keywords 
#2.All data of the web-pages having keywords            
```
